```python
#Profibility of our playoff season predicitions. Uses plus/minus set by NBA to determine how much money was won or lost
def moneylineProfits(df, logreg, wagerpct, starting_sum, limit):
    predictions = logreg.predict_proba(playoff_df[["AST_x", "BLK_x", "DREB_x", "FG3_PCT_x", "FG_PCT_x", "FT_PCT_x", "FTA_x", "OREB_x", "STL_x", "TOV_x"]])
    df_odds = df[['PLUS_MINUS_x', 'WL_x']]
    for index, row in df_odds.iterrows():
        pred = predictions[index]
        if pred[0] > limit:
            wager = wagerpct * starting_sum
            starting_sum = starting_sum - wager
            if row['WL_x'] == 1:
                starting_sum = starting_sum + (wager * 2)
        elif pred[0] < 1 - limit:
            wager = wagerpct * starting_sum
            starting_sum = starting_sum - wager
            if row['WL_x'] == 0:
                starting_sum = starting_sum + (wager * 2)

    return starting_sum
        
```

```python
moneylineProfits(playoff_df, logreg, 0.01, 1000, 0.8)
```
